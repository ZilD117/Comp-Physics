```python
# Rayleigh Oscillator
```

```python
import numpy as np, matplotlib.pylab as plt

# Initial values
t0 = 0
tmax = 100
dt = 0.01
n = int((tmax-t0) / dt)
y = np.zeros(n) 
x = np.zeros(n)
y[0] = 0
x[0] = 0
x[1] = 1e-8

# Constants
m = 2
k = 1
a = 1
b = 1
```

```python
# Equation
for i in range (1, n-1):
    x[i+1] = 2*x[i] - x[i-1] + (dt**2)*(-k*x[i]/m + (a/m)*((x[i]-x[i-1])/dt) - (b/m)*(((x[i]-x[i-1])/dt)**3))
```

```python
# Generate time steps
t = np.arange(t0, tmax, dt)
t = t * dt
print(t)
```

```python
# Plot x vs. t
plt.plot(t, x)
plt.xlabel("t")
plt.ylabel("x")
plt.show()
```

```python
# Define y=x' function
y = np.zeros(n)
for i in range(0, n-1):
    y[i] = (x[i]-x[i+1]) / dt

# Plot v vs. t
plt.plot(t, y)
plt.xlabel("t")
plt.ylabel("y")
plt.show()
```
