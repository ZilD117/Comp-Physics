```python
import numpy as np
from matplotlib import pyplot as plt

# Initial values
n = 1000 # number of sum iterations
x = np.zeros(n) # list of x values
x[0] = 1 # intial x
r = np.linspace(2, 3, 10000) # set of r values
lyapunov = np.zeros(len(r)) # set of lyapunov exponents

# calculate x values from 0 to n for df
def iterateDf(x, n, r):
    for i in range(1, n):
        x[i] = r * np.cos(x[i-1])

# calculate lyapunov exponent for x value
def calcLyapunov(x, n):
    sum = 0
    for i in range(0, n-1):
        sum += np.log(abs(x[i]))
    
    return (1/n) * sum
```

```python
# caluclate lyapunov for list of r values
for i in range(0, len(r)):
    iterateDf(x, n, r[i])
    lyapunov[i] = calcLyapunov(x, n)

# plot lyapunov exponent vs. x
plt.scatter(r, lyapunov, s=0.5, color='b')
plt.axhline(y=0, color='r')
plt.xlabel("r")
plt.ylabel("Lyapunov exponent")
plt.show()
```
