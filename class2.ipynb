\begin{align*}
\ddot{x}+\frac{k}{m} x=\frac{a}{m} \dot{x}-\frac{b}{m} \dot{x}^3 \\
\Rightarrow \frac{x_{i+1}+x_{i-1}-2 x_i}{\Delta t^2}+\frac{k}{m} x_i=\frac{a}{m}\left(\frac{x_{i-}-x_{i-1}}{\Delta t}\right)-\frac{b}{m}\left(\frac{x_i-x_{i-1}}{\Delta t}\right)^3 \\
x_{i+1}+x_{i-1}-2 x_i=\Delta t^2 \cdot\left[-\frac{k}{m} x_i+\frac{a}{m \Delta t} x_i-\frac{a}{m \Delta t} x_{i-1}-\frac{b}{m \Delta t^3} \cdot\left(x_i^3-3 x_i{ }^2 \cdot x_{i-1}+3 x_i x_{i-1}{ }^2-x_{i-1}^3\right)]\right.
\end{align*}

```python
import numpy as np
import matplotlib.pyplot as plt



k = 1
a = 1
b = 1
m = 2

omega = 5
def methods(x0, x1, t0,t_final,delta_t):
    t = np.arange(t0,t_final,delta_t)
    x = np.zeros_like(t)
    x[0] = x0
    x[1] = x1
    for i in range(1,len(x)-1):
        x[i+1] = 2*x[i]-x[i-1]+delta_t**2*(-k/m*x[i]+a/(m*delta_t)*x[i]-a/(m*delta_t)*x[i-1]-b/(m*delta_t**3)*(x[i]**3-3*x[i]**2*x[i-1]+3*x[i]*x[i-1]**2-x[i-1]**3))

    return x,t

# For x-t diagram
x,t = methods(0,1e-8,0,50,0.01)
plt.plot(t,x,label="x-t diagram")
plt.show()


#for v-t diagram & Assume the initial velocity is 0
delta_t = 0.01
v = [(x[i+1]-x[i])/delta_t for i in range(len(x)-1)]
t1 = np.delete(t,0)
plt.plot(t1,v,label="v-t diagram")
plt.legend()
plt.show()



x1 = np.delete(x,0)
plt.plot(x1,v,label="v-x diagram")

plt.legend()
plt.show()
```
