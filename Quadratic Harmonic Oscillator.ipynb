```python
import numpy as np
import matplotlib.pyplot as plt
```

\begin{align*}
m\ddot{x} = F \\
V(x) =\frac{1}{2}kx^2 - \frac{1}{4}kx^4\\
F = - \frac{\partial V}{\partial x} \rightarrow F(x) = -k(x - x^3)\\
F(x) = -\omega^2m(x-x^3)    \because \omega^2 = \frac{k}{m}\\
\ddot{x} = -\omega^2(x-x^3)\\
\frac{x_{i+1} + x_{i-1}-2x_{i}}{\Delta t^2} = -\omega^2(x-x^3)\\
x_{i+1} = -\omega^2(x-x^3)\Delta t^2 - x_{i-1} + 2x_{i}
\end{align*}

```python
w = 5
ti = -5
tf = 5
delta = 0.01
t = np.arange(ti,tf,delta)
x = np.zeros_like(t)

for i in range (1, (len(x)-1)):
    x[0] = 0
    x[1] = 0.01
    x[i+1]=w**2 *(x[i] - x[i]**3 )*delta**2 - x[i-1] + 2*x[i]
plt.plot(t,x)
```

```python
v = np.zeros_like(x)
for i in range (len(x)):
    v[i] = (x[i]-x[i-1])/delta
    v[0]=0
plt.plot(t,v)
```
