```python
import numpy as np, matplotlib.pylab as plt
```

```python
# Equation is xn+1 = rsinxn with 2 < r < 3
# Plot n vs xn+1
# Pick a few different r values and see what happens: make separate plots
```

```python
# Initial values

# r range (given)
rmin = 2
rmax = 3

# n = maximum number of iterations
n = 100

# xvalues holds all of the x(n)
xvalues = np.zeros(n)

# Setting x0
x0 = 1
xvalues[0] = x0

# Equation for our function
def f(x, r):
    return r*np.sin(x)
```

```python
# Iterate through for r = rmin = 2
for i in range(n-1):
    xvalues[i+1] = f(xvalues[i],rmin)
    
# Plot
xaxis = np.linspace(0,n,n)
plt.plot(xaxis,xvalues)
plt.xlabel('n')
plt.ylabel('x(n)')
plt.title('Logistic Map for r = 2 (minimum)')
plt.text(75,xvalues[n-1]-0.1,'x = {}'.format(round(xvalues[n-1],5)))
```

```python
# Iterate through for r = 2.25
for i in range(n-1):
    xvalues[i+1] = f(xvalues[i],2.25)
    
# Plot
plt.plot(xaxis,xvalues)
plt.xlabel('n')
plt.ylabel('x(n)')
plt.title('Logistic Map for r = 2.25')
```
