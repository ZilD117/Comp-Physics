```python
import numpy as np
import matplotlib.pyplot as plt


```

```python


V_0 = -5
alpha = 1
def potential(x):
    return V_0*(np.e**(-alpha*(x**2)))
```

```python
def h1_psi(psi,h, N):
    h_psi=np.zeros(2*N+1)
    wf=[0 for i in range(-N-1,N+2)]

    k=0
    for i in range(-N,N+1):
            wf[i]=psi[k]
            k=k+1
    k=0
    for i in range(-N,N+1):
        h_psi[k]=-0.5*(wf[i+1]+wf[i-1]-2*wf[i])/h**2
        h_psi[k]=h_psi[k]+(potential(i*h))*psi[k]                       
        k=k+1
    return h_psi
```

```python

N = 50
h = 0.1
psi=np.zeros(2*N+1)
hpsi=np.zeros(2*N+1)
dt=0.005

psi[:]=1

#print(psi)
for i in range(1000):
    hpsi=h1_psi(psi,h,N)
#    print(hpsi)
    psi=psi-dt*hpsi
    print(i,np.dot(psi,hpsi)/np.dot(psi,psi))
 
```
