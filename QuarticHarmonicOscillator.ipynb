```python
### Quartic Harmonic Oscillator Finite differentiation
```

```python
import numpy as np, matplotlib.pylab as plt

# Initial values
n = 100000
x = np.zeros(n)
u = np.zeros(n)
v = np.zeros(n)

x[0] = 0
x[1] = 0.0001
v[0] = 0

w = 1
dt = 0.0005

# Equations
for i in range(1, n-1): # position
    x[i+1] = ((dt*w)**2)*(x[i]-(x[i]**3))-x[i-1]+2*x[i]
    
for i in range(0, n-1): # potential
    u[i] = 0.5*(x[i]**2) - 0.25*(x[i]**4)
    
for i in range(1, n-1): # velocity
    v[i+1] = (x[i-1] - x[i]) / dt
```

```python
# Generate time steps
t = np.arange(n)
t = t * dt
print(t)
```

```python
# Plot position vs. time
plt.plot(t, x)
plt.xlabel("t")
plt.ylabel("x")
plt.show()
```

```python
# Plot potential vs. position
plt.plot(x, -u)
plt.xlabel("x")
plt.ylabel("u")
plt.show
```

```python
# Plot velocity vs. position
plt.plot(x, v)
plt.xlabel("x")
plt.ylabel("v")
plt.show
```

```python
# Plot velocity vs. time
plt.plot(t, v)
plt.xlabel("v")
plt.ylabel("t")
plt.show
```
