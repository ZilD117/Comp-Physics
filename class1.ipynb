```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

x = np.linspace(-1.5,1.5,51)
x
```

```python
f = lambda x: -1/2*x**2+1/4*x**4
plt.plot(x,f(x))
plt.xlabel("x")
plt.ylabel("Potential Energy")
```

```python

omega = 5
def methods(x0, x1, t0,t_final,delta_t):
    t = np.arange(t0,t_final,delta_t)
    x = np.zeros_like(t)
    x[0] = x0
    x[1] = x1
    for i in range(1,len(x)-1):
        x[i+1] = delta_t**2*omega**2*(x[i]-x[i]**3)+2*x[i]-x[i-1]
        
        
    return x,t
```

```python
# For x-t diagram
x,t = methods(0,0.01,-5,5,0.01)
plt.plot(t,x)
plt.xlabel("time t")
plt.ylabel("position x")
```

```python
#for v-t diagram & Assume the initial velocity is 0
delta_t = 0.01
v = [(x[i]-x[i-1])/delta_t for i in range(len(x))]
v[0] = 0
plt.plot(t,v)

```

```python
#x-v diagram
plt.plot(x,v)
plt.xlabel("position x")
plt.ylabel("velocity v")
```
