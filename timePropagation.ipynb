```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 50
CONSTANT_A = 1
V_0 = -5
L = 5
x0 = 0
alpha = 1
N = 50
k = 1
dx = delta_x = delta_x = 2*L/n
x = np.linspace(-L,L,2*N)
x_0 = 0
def potential(x):
    return V_0*np.e**(-CONSTANT_A*x**2)

h = np.zeros((n,n))
V = np.zeros(N)
for i in range(0, n):
    V[i] = potential(x[i])

# generate matrix
for i in range(0, n):
    #h[i][i] = (1/(2*(dx**2)))+V[i]
    h[i][i] = 1/(dx**2)+V[i]
    if i != n-1:
        h[i][i+1] = -1/(2*(dx**2))
    if i != 0:
        h[i][i-1] = -1/(2*(dx**2))


psi = np.empty([2*n,1]).astype(complex)
for i in range(0, 2*n):
    psi[i] = np.exp(-alpha * (x[i] - x_0)**2 + 1j * k * x[i])



plt.plot(x,psi.real)
plt.plot(x,psi.imag)

```

```python
N = 50
h = 0.1

dt=0.005
plt.figure(figsize=(12, 6))
plt.subplot(4,1,1)
plt.title("Real part of psi")
plt.plot(x, psi.real, label="Initial psi real")

plt.subplot(4,1,2)
plt.title("Imaginary part of psi")
plt.plot(x, psi.imag, label="Initial psi imag")

plt.subplot(4,1,3)
plt.title("real part of psi **2")
plt.plot(x, psi.real**2, label="Initial psi imag")

plt.subplot(4,1,4)
plt.title("Imaginary part of psi **2")
plt.plot(x, psi.imag**2, label="Initial psi imag")

for i in range(2000):
    
    psi = psi - 1j * np.dot(h, psi) * dt
    if(i%200==0):
        plt.subplot(4,1,1)
        plt.plot(x, psi.real, label=f"Real psi after {i} iterations")
        plt.subplot(4,1,2)
        plt.plot(x, psi.imag, label=f"Imag psi after {i} iterations")
        plt.subplot(4,1,3)
        plt.plot(x, psi.real**2, label=f"real psi**2 after {i} iterations")
        plt.subplot(4,1,4)
        plt.plot(x, psi.imag**2, label=f"imag psi**2 after {i} iterations")
       
```
