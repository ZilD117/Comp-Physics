```python
# Import libraries
import numpy as np
from matplotlib import pyplot as plt
from scipy.stats import norm
```

$V(x) = -V_0 e^{-\alpha x^2}$

```python
n = 100
L = 5
V_0 = -5
alpha = 1
u = np.zeros(n)
u[0] = 1
u[1] = 1
x = np.linspace(-L, L, n)
dx = (2*L)/n
V = np.zeros(n)
h = np.zeros((n, n))

# potential function
def potential(x):
    return V_0*(np.e**(-alpha*(x**2)))

# plot potential
for i in range(0, n):
    V[i] = potential(x[i])

# generate matrix
for i in range(0, n):
    h[i][i] = (1/(dx**2))+V[i]
    if i != n-1:
        h[i][i+1] = -1/(2*(dx**2))
    if i != 0:
        h[i][i-1] = -1/(2*(dx**2))

def plot_u(E):
    for i in range(1, n-1):
        u[i+1] = -2*(dx**2)*(E*u[i] - V_0*u[i]) - u[i-1] + 2*u[i]

    plt.plot(x, u)
```

```python
eig = -3.597582036404873 # eigenvalue
E = eig * np.identity(n)
#E=0
zero_vec = np.zeros(n)

print(h-E)
print(np.linalg.det(h-E))

psi = np.linalg.solve(h-E, zero_vec)
print(psi)

plt.plot(x, psi)
plt.show()
```

```python
E = -3.597582036404873 # eigenvalue
plot_u(E)
E = -3.083747474 # not eigenvalue
plot_u(E)
E = -1.2511981892605217 # eigenvalue
plot_u(E)
plt.show()
```
